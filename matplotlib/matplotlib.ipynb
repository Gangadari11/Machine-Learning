```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[2,4,4,5])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('First plot')
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
plt.plot(x , np.power(x,2))
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Example')
plt.show()
```

```python
x=np.arange(1,10,0.1)
y1=x
y2=np.sqrt(x)
y3=np.power(x,2)
y4=np.power(x,3)
plt.figure()

plt.subplot(2,2,1)
plt.plot(x,y1,'ro')
plt.title('y1=x')


plt.subplot(2,2,2)
plt.plot(x,y2,'g--')
plt.title(r'$y2 = \sqrt{x}$')

plt.subplot(2,2,3)
plt.plot(x,y3,'b^')
plt.title('$y3 = x^2$')

plt.subplot(2,2,4)
plt.plot(x,y4,'ks')
plt.title('$y4 = x^3$')

plt.show()
```
